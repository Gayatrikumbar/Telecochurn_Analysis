# Data Cleaning

```python
telco_data = telco_base_data.copy()
```

#### Missing Value Treatement
###### Since the % of these records compared to total dataset is very low ie 0.15%, it is safe to ignore them from further processing.

```python
#Removing missing values 
telco_data.dropna(how = 'any', inplace = True) # "any" it will drop all the row is any one  value is missing ; "inplace=true" it modifies the dataframe and save changes without creating a new one,but it will default "false"

# telco_data.fillna(0)
```

###### Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on...

```python
# Get the max tenure
print(telco_data['tenure'].max()) 
```

```python
# Group the tenure in bins of 12 months
# Create a list of user-friendly string labels for each 12-month (1-year) tenure group.
# This list will look like: ['1 - 12', '13 - 24', '25 - 36', ...]
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right=False, labels=labels)
"""
Use the pandas 'cut' function to segment the numerical 'tenure' data into bins.
right=False: This specifies that the bins are left-inclusive but right-exclusive. This means a value of 12 months would fall into the '1 - 12' bin, but a value of 13 months would fall into the '13 - 24' bin.
"""
```

```python
telco_data['tenure_group'].value_counts() #selects the new coulum tenure group and  counts the occurence
```

#### Convert all the categorical variables into dummy variables

```python
sns.lmplot(data=telco_data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False) # fit reg is used to hide the regression line drawn default
```

#### Total Charges increase as Monthly Charges increase - as expected.

```python
Mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies["Churn"] == 0) ],
                color="Red", fill = True)
Mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", fill= True) # ax is used to plot both in same graph 
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

##### Insight: Churn is high when Monthly Charges ar high

```python
Tot = sns.kdeplot(telco_data_dummies.TotalCharges[(telco_data_dummies["Churn"] == 0) ],
                color="Red", fill = True)
Tot = sns.kdeplot(telco_data_dummies.TotalCharges[(telco_data_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", fill= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

### **Surprising insight ** as higher Churn at lower Total Charges

###### However if we combine the insights of 3 parameters i.e. Tenure, Monthly Charges & Total Charges then the picture is bit clear :- Higher Monthly Charge at lower tenure results into lower Total Charge. Hence, all these 3 factors viz Higher Monthly Charge, Lower tenure and Lower Total Charge are linkd to High Churn.

```python
def uniplot(df,col,title,hue=None):#defines a function named uniplot with four parameters
    
    sns.set_style('whitegrid')#white background
    sns.set_context('talk')#scaling of plot elements(like lables,lines)
    plt.rcParams["axes.labelsize"] = 20 #font size to 20
    plt.rcParams['axes.titlesize'] = 22 #font size to 22
    plt.rcParams['axes.titlepad'] = 30 #adjust spacing b/w plots
    
    
    temp = pd.Series(data = hue)#adds a tempory series from hue argument
    fig, ax = plt.subplots()#creates a fig and subplot
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())#dynamically calculates the width of the plot based on the number of unique values in the col and hue columns. The extra values (7 and 4) are likely used to ensure the plot is wide enough to prevent labels from overlapping.
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define global blue & orange palette
blue_orange_palette = ["#1f77b4", "#ff7f0e"]  # matplotlib default blue & orange

# Your existing data filtering
new_df1_target0 = telco_data.loc[telco_data["Churn"] == 0]  # Non-churned customers
new_df1_target1 = telco_data.loc[telco_data["Churn"] == 1]  # Churned customers

def uniplot(df, col, title, hue=None):
    """
    Creates a count plot with proper formatting (blue & orange theme)
    """
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 16
    plt.rcParams['axes.titlesize'] = 18
    plt.rcParams['axes.titlepad'] = 20
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    if hue is not None:
        ax = sns.countplot(
            data=df, 
            x=col, 
            hue=hue, 
            order=df[col].value_counts().index,
            palette=blue_orange_palette
        )
    else:
        ax = sns.countplot(
            data=df, 
            x=col, 
            order=df[col].value_counts().index,
            color=blue_orange_palette[0]  # Always blue for single category
        )
    
    plt.xticks(rotation=45, ha='right')
    plt.title(title, pad=20)
    
    for container in ax.containers:
        ax.bar_label(container, fmt='%d')
    
    plt.tight_layout()
    plt.show()

print("=== CHURNED CUSTOMERS ANALYSIS ===")

uniplot(new_df1_target1, col='gender', title='Distribution of Gender for Churned Customers')
uniplot(new_df1_target1, col='Partner', title='Distribution of Partner Status for Churned Customers', hue='gender')
uniplot(new_df1_target1, col='PaymentMethod', title='Distribution of Payment Method for Churned Customers', hue='gender')
uniplot(new_df1_target1, col='Contract', title='Distribution of Contract Type for Churned Customers', hue='gender')
uniplot(new_df1_target1, col='TechSupport', title='Distribution of Tech Support for Churned Customers', hue='gender')
uniplot(new_df1_target1, col='SeniorCitizen', title='Distribution of Senior Citizen Status for Churned Customers', hue='gender')

print("\n=== NON-CHURNED CUSTOMERS ANALYSIS ===")
uniplot(new_df1_target0, col='Partner', title='Distribution of Partner Status for Non-Churned Customers', hue='gender')

def comparison_plot(df, col, title):
    """
    Creates side-by-side comparison of churned vs non-churned customers
    (blue & orange theme)
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    churned = df[df['Churn'] == 1]
    sns.countplot(data=churned, x=col, hue='gender', ax=ax1, palette=blue_orange_palette)
    ax1.set_title(f'{title} - Churned Customers')
    ax1.tick_params(axis='x', rotation=45)
    
    not_churned = df[df['Churn'] == 0]
    sns.countplot(data=not_churned, x=col, hue='gender', ax=ax2, palette=blue_orange_palette)
    ax2.set_title(f'{title} - Non-Churned Customers')
    ax2.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()

# Example comparison plot
comparison_plot(telco_data, 'PaymentMethod', 'Payment Method Distribution')
```
